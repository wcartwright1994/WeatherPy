```python
# Dependencies and Setup
import pandas as pd
import requests
import gmaps
import numpy as np

# Import API key
from config import api_key_goog as g_key

# Configure gmaps API key
gmaps.configure(api_key=g_key)
```

```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv(r'C:\Users\helpdesk\Documents\Data Analytics Bootcamp\Modules\Module 6\WeatherPy_Database.csv')
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 
min_temp = float(input('What is your minimum desired temperature?'))
max_temp = float(input('What is your maximum desired temperature?'))
```

```python
type(min_temp)
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.
city_data_df = city_data_df.rename(columns = {'Max Temp':'Max_Temp','Wind Speed':'Wind_Speed','Current Description':'Current_Description'})

temp_df = city_data_df[(city_data_df.Max_Temp >= min_temp) & (city_data_df.Max_Temp <= max_temp)]
temp_df
```

```python
# 4a. Determine if there are any empty rows.
temp_df.isnull().values.any()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesn’t have empty rows.
clean_df = temp_df.dropna(axis='rows')
clean_df
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = clean_df[["City", "Country", "Max_Temp", "Current_Description", "Lat", "Lng"]].copy()

# 5b. Create a new column "Hotel Name"
hotel_df["Hotel_Name"] = ""
hotel_df.head(10)
```

```python
# 6a. Set parameters to search for hotels with 5000 meters.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"   
    
# 6b. Iterate through the hotel DataFrame.

for index, row in hotel_df.iterrows():

    
    # 6c. Get latitude and longitude from DataFrame
    lat_city = row['Lat']
    lng_city = row['Lng']
    location = f'{lat_city},{lng_city}'
    
    current_city = row['City']
    params['location'] = location
    
    # 6d. Set up the base URL for the Google Directions API to get JSON data.
    response = requests.get(base_url,params=params).json()

    # 6e. Make request and retrieve the JSON data from the search. 
    results = response['results']

    # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
    try:
        hotel_df.loc[index, 'Hotel_Name'] = results[0]['name']

    except (KeyError, IndexError):
        print(f'No hotel found in {current_city}.')
        hotel_df.loc[index, 'Hotel_Name'] = np.nan
        
print('Hotel search complete.')
```
